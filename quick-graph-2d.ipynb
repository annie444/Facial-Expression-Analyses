```python
with pd.HDFStore(
        "/Users/annieehler/Projects/Jupyter_Notebooks/python_outputs/data.h5"
) as data_file:
    data = data_file.get('all_trial_data')

    data = data.dropna(axis=0, thresh=4, subset=["upper_eye_x", "upper_eye_y", "lower_eye_x", "lower_eye_y", "upper_ear_x", "upper_ear_y", "lower_ear_x", "lower_ear_y", "outer_ear_x", "outer_ear_y", "upper_whisker_x", "upper_whisker_y", "outer_whisker_x", "outer_whisker_y", "lower_whisker_x", "lower_whisker_y", "upper_mouth_x", "upper_mouth_y", "outer_mouth_x", "outer_mouth_y", "lower_mouth_x", "lower_mouth_y", "inner_nostril_x", "inner_nostril_y", "outer_nostril_x", "outer_nostril_y"])

    weeks = [0, 1, 5, 6]

    trials = ["Sucrose", "Airpuff"]

    new_data = []

    for week in weeks:
        df = data[(data.week_list == week)]
        for trial in trials:
            df1 = df[(df.trial_type == trial)]
            mice = df1.mouse_list.unique().tolist()
            for mouse in mice:
                df2 = df1[(df1.mouse_list == mouse)]
                ind_trials = df2.trial_num.unique().tolist()
                for ind_trial in ind_trials:
                    df3 = df2[(df2.trial_num == ind_trial)]
                    new_data.append(df3)


print(len(new_data))
print(new_data[0])
print(new_data[5])
```

```python
for i, data in enumerate(new_data):
    min_frame = data.frame_list.min()
    data["frame_trial_idx"] = data["frame_list"] - min_frame
    new_data[i] = data

print(len(new_data))
print(new_data[0])
print(new_data[5])
```

```python
data = pd.concat(new_data)

cse_array = ['CSE008', 'CSE020']
csc_array = ['CSC009', 'CSC013']

cse = data.mouse_list.isin(cse_array)
csc = data.mouse_list.isin(csc_array)

data_cse = data[cse]
data_csc = data[csc]

data_cse
```
