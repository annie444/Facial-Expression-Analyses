```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import date
```

```python
with pd.HDFStore(
        "/Users/annieehler/Projects/Jupyter_Notebooks/python_outputs/exprs.h5"
) as data_file:
    data = data_file.get('/SLEAP_data_by_trial/D3_PCA_trial_data')

    data = data.dropna(axis=0, thresh=2, subset=["principal component 1", "principal component 2"])

    weeks = [0, 1, 5, 6]

    trials = ["Sucrose", "Airpuff"]

    new_data = []

    for week in weeks:
        df = data[(data.week_list == week)]
        for trial in trials:
            df1 = df[(df.trial_type == trial)]
            mice = df1.mouse_list.unique().tolist()
            for mouse in mice:
                df2 = df1[(df1.mouse_list == mouse)]
                ind_trials = df2.trial_num.unique().tolist()
                for ind_trial in ind_trials:
                    df3 = df2[(df2.trial_num == ind_trial)]
                    new_data.append(df3)


print(len(new_data))
print(new_data[0])
print(new_data[5])
```

```python
drop_data = []

for data in new_data:
    data = data.drop(["level_0", "level_1", "level_2", "level_3"], axis=1)
    drop_data.append(data)

print(len(drop_data))
print(drop_data[0])
print(drop_data[5])
```

```python
new_data = []

for data in drop_data:
    min_frame = data.frame_list.min()
    data["frame_trial_idx"] = data.frame_list - min_frame
    new_data.append(data)

print(len(new_data))
print(new_data[0])
print(new_data[5])
```

```python
data = pd.concat(new_data)

cse_array = ['CSE008', 'CSE020']
csc_array = ['CSC009', 'CSC013']

cse = data.mouse_list.isin(cse_array)
csc = data.mouse_list.isin(csc_array)

data_cse = data[cse]
data_csc = data[csc]

data_cse
```

```python
mean_and_std = []

for week in weeks:
    for trial in trials:
        df = data_cse[(data_cse.week_list == week) & (data_cse.trial_type == trial)]
        mean_and_std.append(df.groupby(by=["week_list", "trial_type", "frame_trial_idx"]).agg({"principal component 1": ['mean', 'std'], "principal component 2": ['mean', 'std'], "principal component 3": ['mean', 'std']}))

mean_and_std_df_cse = pd.concat(mean_and_std)

mean_and_std_df_cse
```

```python
mean_and_std = []

for week in weeks:
    for trial in trials:
        df = data_csc[(data_csc.week_list == week) & (data_csc.trial_type == trial)]
        mean_and_std.append(df.groupby(by=["week_list", "trial_type", "frame_trial_idx"]).agg({"principal component 1": ['mean', 'std'], "principal component 2": ['mean', 'std'], "principal component 3": ['mean', 'std']}))

mean_and_std_df_csc = pd.concat(mean_and_std)

mean_and_std_df_csc
```

```python
diff = data_cse[["timestamps", "principal component 1", "principal component 2"]].diff()
locational_vector = np.sqrt((diff["principal component 1"] ** 2 ) + (diff["principal component 2"] ** 2))
velocity = locational_vector / diff["timestamps"]
acceleration = velocity / diff["timestamps"]

data_cse["locational_vector"] = locational_vector
data_cse["velocity"] = velocity
data_cse["acceleration"] = acceleration

data_cse.dropna(axis=0, subset=["locational_vector", "velocity", "acceleration"])
```

```python
diff = data_csc[["timestamps", "principal component 1", "principal component 2"]].diff()
locational_vector = np.sqrt((diff["principal component 1"] ** 2 ) + (diff["principal component 2"] ** 2))
velocity = locational_vector / diff["timestamps"]
acceleration = velocity / diff["timestamps"]

data_csc["locational_vector"] = locational_vector
data_csc["velocity"] = velocity
data_csc["acceleration"] = acceleration

data_csc.dropna(axis=0, subset=["locational_vector", "velocity", "acceleration"])
```

```python
mean_and_std = []

for week in weeks:
    for trial in trials:
        df = data_cse[(data_cse.week_list == week) & (data_cse.trial_type == trial)]
        mean_and_std.append(df.groupby(by=["week_list", "trial_type", "frame_trial_idx"]).agg({"locational_vector": ['mean', 'std'], "velocity": ['mean', 'std'], "acceleration": ['mean', 'std']}))

mean_and_std_df_cse = pd.concat(mean_and_std)

mean_and_std_df_cse["locational_vector", "upper_std"] = mean_and_std_df_cse["locational_vector", "mean"]+(mean_and_std_df_cse["locational_vector", "std"] * 2)
mean_and_std_df_cse["locational_vector", "lower_std"] = mean_and_std_df_cse["locational_vector", "mean"]-(mean_and_std_df_cse["locational_vector", "std"] * 2)
mean_and_std_df_cse["velocity", "upper_std"] = mean_and_std_df_cse["velocity", "mean"]+(mean_and_std_df_cse["velocity", "std"] * 2)
mean_and_std_df_cse["velocity", "lower_std"] = mean_and_std_df_cse["velocity", "mean"]-(mean_and_std_df_cse["velocity", "std"] * 2)
mean_and_std_df_cse["acceleration", "upper_std"] = mean_and_std_df_cse["acceleration", "mean"]+(mean_and_std_df_cse["acceleration", "std"] * 2)
mean_and_std_df_cse["acceleration", "lower_std"] = mean_and_std_df_cse["acceleration", "mean"]-(mean_and_std_df_cse["acceleration", "std"] * 2)
mean_and_std_df_cse = mean_and_std_df_cse.sort_index(axis=1)
mean_and_std_df_cse
```
